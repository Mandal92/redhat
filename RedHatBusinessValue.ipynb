```python
def preprocess_features(data):
    
    cols = list(data.columns)
    bools = cols[23:-2]
    
    #We are running loop 2 times as outcome is there in the column 13 which is int
    for col in cols[2:13]:
        data[col] = data[col].fillna('type 0')
        data[col] = data[col].apply(lambda x: x.split(' ')[1])
        data[col] = pd.to_numeric(data[col]).astype(int)
    
    for col in cols[14:23]:
        data[col] = data[col].fillna('type 0')
        data[col] = data[col].apply(lambda x: x.split(' ')[1])
        data[col] = pd.to_numeric(data[col]).astype(int)
        
    for col in bools:
        data[col] = pd.to_numeric(data[col]).astype(int)
        
    data['people_id'] = data['people_id'].apply(lambda x: x.split('_')[1])
    data['people_id'] = pd.to_numeric(data['people_id']).astype(int)
    
    data['activity_id'] = data['activity_id'].apply(lambda x: x.split('_')[1])
    data['activity_id'] = pd.to_numeric(data['activity_id']).astype(int)
    return data
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
```

```python
model = XGBClassifier(n_jobs=-1, n_estimators=100, learning_rate=0.05, max_depth=5)
```

```python
y = sample.iloc[:,12].values
y
```

```python
input = sample.drop('outcome', axis=1)
```

```python
X = input.iloc[:,2:].values
X
```

```python
test = features.drop('outcome', axis=1)
```

```python
X_test = test.iloc[:,2:].values
X_test
```

```python
y_pred = model.predict(X_test)
```

```python
predictions = [round(value) for value in y_pred]
```
